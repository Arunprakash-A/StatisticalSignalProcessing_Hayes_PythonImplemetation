 * Ensemble averages are required for Signal Modelling, optimum filtering and spectrum estimation. However, oftentimes we have a very few realizations (or left with a single realization) to estimate these averages.
 * Can we use the time average with a single realization? What are all the conditions?
 * If the time average of a process equals the ensemble average then we call it an ergodic process.

**Convergence**
 
  1. A RP, $x(n)=A$ where $A$ is RV taking value 1 or -1 with $P(A=1 \ or \ -1) = 0.5$. Will time avergage converges to ensemble averages?
  2. Will time avergage converges to ensemble averages for a **Bernoulli Process**?. $x(n)=1$ if head, $x(n)=-1$ if tails turn out  at time instant $n$.

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import signal as sig
```

```python
A = np.random.choice([1,-1],size=4)
step = np.ones(20)
fig,ax = plt.subplots(4,1,figsize=(10,8))
for i in range(4):
    ax[i].stem(A[i]*step,use_line_collection=True)
    ax[i].axis([-1,20,-2,2])
    ax[i].grid()
ax[0].set_title('Sample Realizations')
ax[3].set_xlabel('index n')
fig.tight_layout() 
```

 * Therefore, the mean of the process is $E\{x(n)\}=E\{A\}=0$. 

 * However, if we compute the time average over $N$ elements in a realization, then $m_x(N)=1$ or $m_x(N)=-1$ with probability of 0.5 **but will never converge to the actual mean if $N \to \infty$** as $r_x(k)=1$ (Observe, it is a constant, not a function of $k-l$). 
 
 *  Therefore, adding more sample values do not bring any new information.
 * On the other hand, if we consider a Bernoulli process (as we have discussed in the earlier notebook), the time average of the process is given by $$ \frac{1}{N} \sum_{n=0}^{N-1} x(n) = \frac{n_1}{N} - \frac{n_2}{N}$$
 
 where $n_1$ is the number of times head appeared and $n_2$ is the number of times tail appeared.
 
 * As the sample values increases ($N \to \infty$) it adds new information and converges to the original mean (**The bernoulli process contains uncorrelated RVs, therefore each new sample improves the estimate**)
 
 * Clearly, Bernoulli Process is **Ergodic**. Then the condition for convergence is formulated in the MSE sense as 
 $$ \lim_{N\to\infty} E\bigg\{[m_x(N)-m_x]^2\bigg\}=0$$
